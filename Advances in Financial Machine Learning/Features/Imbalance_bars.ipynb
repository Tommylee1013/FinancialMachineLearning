```python
plt.figure(figsize = (10, 4))
plt.scatter(volume_imbalance['date_time'], volume_imbalance.close, s = 1, marker = '.', edgecolor = 'none')
plt.grid(False)
plt.title('Volume Imbalance Bar')
plt.show()
```

#### Dollar imbalance bar

```python
dollar_imbalance,_ = bar.const_dollar_imbalance_bar(expected_imbalance_window = 300, exp_num_ticks_init = 100)
dollar_imbalance
```

```python
plt.figure(figsize = (10, 4))
plt.scatter(dollar_imbalance['date_time'], dollar_imbalance.close, s = 1, marker = '.', edgecolor = 'none')
plt.grid(False)
plt.title('Dollar Imbalance Bar')
plt.show()
```

```python
tib_returns = tick_imbalance.close.pct_change().dropna()
vib_returns = volume_imbalance.close.pct_change().dropna()
dib_returns = dollar_imbalance.close.pct_change().dropna()

tib_returns.index = tick_imbalance.date_time[1:]
vib_returns.index = volume_imbalance.date_time[1:]
dib_returns.index = dollar_imbalance.date_time[1:]
```

```python
tib_standard = (tib_returns - tib_returns.mean())/tib_returns.std()
vib_standard = (vib_returns - vib_returns.mean())/vib_returns.std()
dib_standard = (dib_returns - dib_returns.mean())/dib_returns.std()
```

```python
plt.figure(figsize = (10, 4))
plt.scatter(tib_standard.index, tib_standard, s = 2, marker = '.', label = 'Tick Imbalance', alpha = 0.5, edgecolor = 'none')
plt.scatter(vib_standard.index, vib_standard, s = 2, marker = '.', label = 'Volume Imbalance', alpha = 0.5, edgecolor = 'none')
plt.scatter(dib_standard.index, dib_standard, s = 2, marker = '.', label = 'Dollar Imbalance', alpha = 0.5, edgecolor = 'none')
plt.legend()
plt.grid(False)
plt.title('Standardized Returns')
plt.show()
```

```python
import seaborn as sns
sns.kdeplot(tib_standard, label = 'Tick Imbalance', color = 'blue')
sns.kdeplot(vib_standard, label = 'Volume Imbalance', color = 'red')
sns.kdeplot(dib_standard, label = 'Dollar Imbalance', color = 'green')
sns.kdeplot(np.random.normal(size=1000000), label = "Normal distribution", color = 'black')
plt.title('Standardized Returns')
plt.grid(False)
plt.xlim([-5, 5])
plt.legend()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, ax = plt.subplots(figsize = (7, 4))
plot_pacf(np.log(tick_imbalance['close']), auto_ylims = True, lags = 20, ax = ax)
plt.xlabel('Time lags', fontsize = 10)
plt.title('PACF of tick imbalance bar')
plt.grid(False)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (7, 4))
plot_acf(
    (np.log(tick_imbalance['close']) - np.log(tick_imbalance['close'].shift(1))).dropna(), 
    auto_ylims = True, lags = 20, ax = ax
)
plt.grid(False)
plt.xlabel('Time lags', fontsize = 10)
plt.title('ACF of tick imbalance bar')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (7, 4))
plot_pacf(np.log(time_5), auto_ylims = True, lags = 20, ax = ax)
plt.xlabel('Time lags', fontsize = 10)
plt.title('PACF of time bar')
plt.grid(False)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (7, 4))
plot_acf(
    (np.log(time_5) - np.log(time_5.shift(1))).dropna(), 
    auto_ylims = True, lags = 20, ax = ax
)
plt.grid(False)
plt.xlabel('Time lags', fontsize = 10)
plt.title('ACF of time bar')
plt.show()
```
