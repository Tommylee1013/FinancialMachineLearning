```python
dollar = pd.read_csv('./Data/dollar_bar.csv', index_col = 0, parse_dates = True)
```

### Heteroscedasticity models

#### 1. ARCH model

```python
models = HeteroscedasticityModels(dollar.close, vol = 'original')
```

```python
result = models.arch(p = 6)
result
```

```python
result.conditional_volatility.plot(grid = False, figsize = (10,4))
plt.title('Extimated Volatility with ARCH model', fontsize = 12, fontfamily = 'Serif')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$\sigma$')
plt.ylim([0,0.03])
```

#### 2. GARCH model

```python
result = models.garch(p = 2, q = 2)
result
```

```python
result.conditional_volatility.plot(grid = False, figsize = (10, 4))
plt.title('Extimated Volatility with GARCH model', fontsize = 12, fontfamily = 'Serif')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$\sigma$')
plt.ylim([0,0.03])
```
