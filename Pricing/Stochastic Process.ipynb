### Stochastic Process

```python
from FinancialMachineLearning.generator.stochasticprocess import MonteCarloSimulation
```

```python
simulation = MonteCarloSimulation(
    0, initial_price = 1000, maturity = 1, sigma = 0.2, 
    dividend_yield = 0, nObs = 100, slices = 10, random_state = False
)
```

### Vasicek Model

```python
vasicek = simulation.vasicek_model(kappa = 0.1, theta = 0.01, sigma_r = 0.01)
```

```python
import matplotlib.pyplot as plt

plt.plot(vasicek.cumsum(axis = 0), color = 'darkorange', alpha = 0.7)
plt.title('Vasicek model simulation')
plt.show()
```

### Cox Ingersoll Ross Model

```python
cir = simulation.cox_ingersoll_ross_model(kappa = 0.5, theta = 0.001, sigma_r = 0.5)
```

```python
plt.plot(cir.cumsum(axis = 0), color = 'darkorange', alpha = 0.7)
plt.title('Cox ingersoll ross model simulation')
plt.show()
```

### Heston Model

```python
heston = simulation.heston_model(kappa = 0.2, theta = 0.01, sigma_v = 0.1, rho = 0.5)
```

```python
plt.plot(heston, color = 'darkorange', alpha = 0.7)
plt.title('Heston model simulation')
plt.show()
```

```python
plt.hist(heston.mean(axis = 1), bins = 100, color = 'darkorange')
plt.show()
```

### Geometric Brownian motion

```python
gbm = simulation.geometric_brownian_motion()
```

```python
plt.plot(gbm, color = 'darkorange', alpha = 0.7)
plt.title('Geometric Brownian Motion simulation')
plt.show()
```

#### Ornstein Uhlenbeck Process

```python
ou = simulation.ornstein_uhlenbeck(kappa = 0.0001, theta = 0.002)
```

```python
plt.plot(ou, color = 'darkorange', alpha = 0.7)
plt.title('Ornstein Uhlenbeck simulation')
plt.show()
```

#### Jump Diffusion Model

```python
jump = simulation.jump_diffusion_model(jump_alpha = 0.1, jump_std = 0.1, poisson_lambda = 0.001)
```

```python
plt.plot(jump, color = 'darkorange', alpha = 0.7)
plt.title('Jump diffusion simulation')
plt.show()
```
