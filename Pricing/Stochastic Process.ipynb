### Stochastic Process

```python
from FinancialMachineLearning.generator.stochasticprocess import MonteCarloSimulation
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
simulation = MonteCarloSimulation(
    interest_rate = 0.05,    
    initial_price = 100,     
    maturity = 0.5,          
    sigma = 0.2,           
    dividend_yield = 0.02,  
    nObs = 1000,           
    slices = 504,       
    random_state = 42 
)
```

### Vasicek Model

```python
vasicek_paths = simulation.vasicek_model(kappa=0.1, theta=0.05, sigma_r=0.02)
plt.plot(vasicek_paths.T, color = 'darkorange', alpha = 0.03)
plt.title('Vasicek Model')
plt.show()
```

### Cox Ingersoll Ross Model

```python
cir_paths = simulation.cox_ingersoll_ross_model(kappa = 0.15, theta = 0.03, sigma_r = 0.01)
plt.plot(cir_paths.T, color = 'darkorange', alpha = 0.03)
plt.title('Cox-Ingersoll-Ross Model')
plt.show()
```

### Heston Model

```python
heston_paths = simulation.heston_model(kappa=0.1, theta=0.04, sigma_v=0.1, rho=-0.7)
plt.plot(heston_paths.T, color = 'darkorange', alpha = 0.03)
plt.title('Heston Model')
plt.show()
```

### Geometric Brownian motion

```python
gbm_paths = simulation.geometric_brownian_motion()
plt.plot(gbm_paths.T, color = 'darkorange', alpha = 0.03)
plt.title('Geometric Brownian Motion')
plt.show()
```

#### Ornstein Uhlenbeck Process

```python
ou_paths = simulation.ornstein_uhlenbeck(kappa = 0.15, theta = 105, sigma = 2)
plt.plot(ou_paths.T, color = 'darkorange', alpha = 0.03)
plt.title('Ornstein-Uhlenbeck Model')
plt.show()
```

#### Jump Diffusion Model

```python
jd_paths = simulation.jump_diffusion_model(jump_intensity=0.2, jump_mean=-0.05, jump_std=0.1)
plt.plot(jd_paths.T, color = 'darkorange', alpha = 0.03)
plt.title('Jump Diffusion Model')
plt.show()
```
