### Deflated Sharpe Ratio

```python
from tqdm import tqdm
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skewnorm, norm
from FinancialMachineLearning.backtest.backtest_statistics import *
```

```python
np.random.seed(42)
```

```python
etf_ret = pd.read_csv('../../Data/ETFs_returns_2019-2020.csv', index_col = 0)
etf_ret.index = pd.to_datetime(etf_ret.index)
universe = etf_ret.columns
```

```python
def generate_random_weights(ret):
    l = [0.01, 0.25, 0.5, 0.75, 1]  # do not use 0, for avoiding zero division
    
    random_weights = np.random.choice(l, size = (ret.shape), replace = True)
    random_weights = np.divide(random_weights, random_weights.sum(axis = 1)[:, None])
        
    weights_df = pd.DataFrame(
        random_weights, 
        index = ret.index,
        columns = ret.columns
    )
    
    return weights_df

def simulate(ret, weights):
    assets_weights = weights.ffill()
    pf_returns = ret.add(1).mul(assets_weights.shift(1)).sum(axis=1).sub(1)
    pf_returns.iloc[0] = 0  # first day return is 0, because we do not have weights for yesterday
    return pf_returns

def returns_to_equity(ret):
    equity = ret.add(1).cumprod().sub(1)
    return equity
```

```python
import warnings
warnings.filterwarnings('ignore')

M = 5000

df_pfs_returns = pd.DataFrame()

for i in tqdm(range(M)):
    _weights = generate_random_weights(etf_ret)
    
    _pf_returns = simulate(etf_ret, _weights)
    
    df_pfs_returns[i] = _pf_returns
```

```python
returns_to_equity(
    df_pfs_returns.sample(n = 10, axis = 1)
).plot(figsize = (8, 5), legend = True)
plt.grid(False)
plt.title("ETF return simulation")
plt.show()
```

```python
ann_best_srs = ann_estimated_sharpe_ratio(df_pfs_returns).sort_values(ascending = False)
ann_best_srs
```

```python
psr_series = pd.Series(dtype = float)

for i in range(len(df_pfs_returns.columns)) :
    psr = probabilistic_sharpe_ratio(
            observed_sr =  df_pfs_returns[[i]].mean() / df_pfs_returns[[i]].std(),
            benchmark_sr = 0,
            number_of_returns = len(df_pfs_returns[[i]]), 
            skewness_of_returns = df_pfs_returns[[i]].skew(),
            kurtosis_of_returns = df_pfs_returns[[i]].kurt()
        )
    psr_series.loc[i] = psr[0]
```

```python
psr_series.sort_values(ascending = False)
```

Analyze results with Deflated Sharpe Ratio

```python
best_psr_pf_returns = df_pfs_returns[psr_series.sort_values(ascending = False).index[0]]
```

```python
idx = psr_series.sort_values(ascending = False).index[0]

dsr = deflated_sharpe_ratio(
    observed_sr = df_pfs_returns[[idx]].mean() / df_pfs_returns[[idx]].std(),
    sr_estimates = (psr_series * 0.1).tolist(),
    number_of_returns = len(df_pfs_returns[[idx]]),
    skewness_of_returns = df_pfs_returns[[idx]].skew(),
    kurtosis_of_returns = df_pfs_returns[[idx]].kurt()
)
```
