### Backtesting with synthetic data

```python
import numpy as np
import pandas as pd
from random import gauss
from itertools import product
from tqdm import tqdm
from FinancialMachineLearning.backtest.backtest_synthetic_data import *
```

#### 1. 제로 장기 균형의 경우

forecast = 0, half life = 5, sigma = 1

```python
rPT = rSLm = np.linspace(0, 10, 21)
coeffs = {'forecast' : 0, 'hl' : 5, 'sigma' : 1}

output = synthetic_simulation(
    coeffs = coeffs, 
    nIter = 1e5, 
    maxHP = 100, 
    rPT = rPT, 
    rSLm = rSLm
)
```

```python
sr_test = get_sharpe_grid(output, rPT, rSLm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(sr_test)
plt.title(f"forecast = {coeffs['forecast']} | half-life = {coeffs['hl']}  | sigma = {coeffs['sigma']}")
plt.xlabel("Profit taking")
plt.ylabel("Stop loss")
plt.grid(False)
plt.show()
```
