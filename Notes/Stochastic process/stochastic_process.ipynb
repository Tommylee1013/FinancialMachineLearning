### Stochastic Process

```python
from FinancialMachineLearning.generator.stochasticprocess import *
```

#### Ornstein Uhlenbeck Process

$$dS = \alpha (\mu - S)dt + \sigma dW$$

```python
ornstein_uhlenbeck = OrnsteinUhlenbeckProcess(alpha = -0.001, mu = 0.02, 
                                              sigma = 0.03, n_paths = 1000, 
                                              n_steps = 1000, start = 10, 
                                              end = 252, initial_price = 1)
```

```python
ornstein_uhlenbeck.mean()
```

```python
ornstein_uhlenbeck.var()
```

```python
simulation = ornstein_uhlenbeck.simulate()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 4))
plt.plot(simulation, color = 'red', alpha = 0.02)
plt.title('Ornstein Uhlenbeck Process', fontsize = 12, fontfamily = 'Serif')
```

#### Geometric Brownian Motion

$$dS = \mu S dt + \sigma S dW$$

```python
brownian_motion = GeometricBrownianMotion(mu = 0, sigma = 0.028, 
                                          n_paths = 1000, n_steps = 1000, 
                                          start = 10, end = 252, initial_price = 1)
```

```python
brownian_motion.mean()
```

```python
brownian_motion.var()
```

```python
simulation = brownian_motion.simulate()
```

```python
plt.figure(figsize = (10, 4))
plt.plot(simulation, color = 'green', alpha = 0.02)
plt.title('Geometric Brownian Motion', fontsize = 12, fontfamily = 'Serif')
```

#### Jump Diffusion Process

$$S = (\mu - 0.5\sigma^2)t + \sigma\cdot W(t) + \sum_{i=1}^{N(t)} Z_i$$

```python
jump_diffusion = JumpDiffusionProcess(mu = 0, sigma = 0.028, lambdaN = 0.00001, 
                                      eta1 = 0.02, eta2 = 0.04, p = 0.4,
                                      n_paths = 1000, n_steps = 1000,
                                      start = 10, end = 252, initial_price = 1)
```

```python
jump_diffusion.mean()
```

```python
jump_diffusion.var()
```

```python
simulation = jump_diffusion.simulate()
```

```python
plt.figure(figsize = (10, 4))
plt.plot(simulation, color = 'blue', alpha = 0.02)
plt.title('Jump Diffusion Model', fontsize = 12, fontfamily = 'Serif')
```

#### Auto Regressive Model

$$r_t = \alpha_0 + \sum_{i=1}^p \alpha_i r_{t-i} + \epsilon_t$$

```python
auto_regressive = AutoRegressiveProcess(p = 3, n_paths = 1000, n_steps = 1000, 
                                        start = 10, end = 252, initial_price = 1)
```

```python
auto_regressive.mean()
```

```python
auto_regressive.var()
```

```python
simulation = auto_regressive.simulate()
```

```python
plt.figure(figsize = (10, 4))
plt.plot(simulation.cumsum(), color = 'blue', alpha = 0.02)
plt.title('Auto Regressive Model', fontsize = 12, fontfamily = 'Serif')
```

#### de Prado Synthetic process

$$S = \mu t + \sigma \cdot W(t) + \sum_{i=1}^{N(t)}Z_i$$

```python
prado = PradoSyntheticProcess(n_features = 1000, n_informative = 100, n_redundant = 100, n_samples = 1000)
```

```python
prado.mean()
```

```python
prado.var()
```

```python
test = prado.simulate()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 4))
plt.plot(test.cumsum(), color = 'purple', alpha = 0.02)
plt.title('de Prado Synthetic Process', fontsize = 12, fontfamily = 'Serif')
```
