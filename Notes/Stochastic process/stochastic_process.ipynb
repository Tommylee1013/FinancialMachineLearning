### Stochastic Process

```python
from FinancialMachineLearning.generator.stochasticprocess import *
```

#### Ornstein Uhlenbeck Process

```python
ornstein_uhlenbeck = OrnsteinUhlenbeckProcess(alpha = -0.001, mu = 0.02, sigma = 0.03, n_paths = 1000, n_steps = 1000, t = 10, T = 252, S_0 = 1)
```

```python
ornstein_uhlenbeck.mean()
```

```python
ornstein_uhlenbeck.var()
```

```python
simulation = ornstein_uhlenbeck.simulate()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 4))
plt.plot(simulation, color = 'red', alpha = 0.02)
plt.title('Ornstein Uhlenbeck Process', fontsize = 12, fontfamily = 'Serif')
```

#### Geometric Brownian Motion

```python
brownian_motion = GeometricBrownianMotion(mu = 0, sigma = 0.028, n_paths = 1000, n_steps = 1000, t = 10, T = 252, S_0 = 1)
```

```python
brownian_motion.mean()
```

```python
brownian_motion.var()
```

#### Jump Diffusion Process

```python
jump_diffusion = JumpDiffusionProcess(mu = 0, sigma = 0.028, lambdaN = 0.00001, eta1 = 0.02, eta2 = 0.04, p = 0.4,
                                      n_paths = 1000, n_steps = 1000, t = 10, T = 252, S_0 = 1)
```

```python
jump_diffusion.mean()
```

```python
jump_diffusion.var()
```

```python
simulation = jump_diffusion.simulate()
```

```python
plt.figure(figsize = (10, 4))
plt.plot(simulation, color = 'blue', alpha = 0.02)
plt.title('Jump Diffusion Model', fontsize = 12, fontfamily = 'Serif')
```

#### Auto Regressive Model

```python
auto_regressive = AutoRegressiveProcess(p = 3, n_paths = 50, n_steps = 10, t = 10, T = 252, S_0 = 1)
```

```python
auto_regressive.mean()
```

```python
auto_regressive.var()
```

```python
simulation = auto_regressive.simulate()
```

```python
plt.figure(figsize = (10, 4))
plt.plot(simulation.cumsum(), color = 'blue', alpha = 0.2)
plt.title('Auto Regressive Model', fontsize = 12, fontfamily = 'Serif')
```
