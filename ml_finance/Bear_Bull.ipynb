### Bull-Bear Market Research

```python
import yfinance as yf

start = '2010-01-01'
end = '2024-03-02'
ticker = 'AAPL'
data = yf.download(ticker, start, end)
```

```python
data['Buy Pressure'] = (data['High'] - data['Open']) / data['Open']
data['Sell Pressure'] = (data['Open'] - data['Low']) / data['Open']
```

```python
data
```

```python
import matplotlib.pyplot as plt

plt.hist(
    data['Buy Pressure'], 
    bins = 100, 
    edgecolor = 'none'
)
plt.hist(
    data['Sell Pressure'], 
    bins = 100, 
    edgecolor = 'none',
    alpha = 0.5
)
plt.grid(False)
plt.show()
```

```python
plt.hist(data['Buy Pressure'] - data['Sell Pressure'], bins = 100, edgecolor = 'none')
plt.grid(False)
plt.show()
```

```python
plt.plot((data['Buy Pressure'] - data['Sell Pressure']) * data['Volume'])
plt.grid(False)
plt.show()
```

```python
plt.hist((data['Buy Pressure'] - data['Sell Pressure']) * data['Volume'], bins = 10000, edgecolor = 'none')
plt.grid(False)
plt.show()
```
